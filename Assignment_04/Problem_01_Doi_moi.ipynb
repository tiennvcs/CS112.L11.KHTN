### Tóm tắt bài toán
-----

Cho dãy A tăng dần (phần tử đầu tiên có chỉ mục là 1) được xác định bởi công thức $A[i] = (a+b)*x - ((i//k)*a + (i//m)*b)$, tìm giá trị $i$ **nhỏ nhất** trong dãy sao cho $A[i] \ge n$.

### Nhận diện khuôn mẫu
----

- Tìm kiếm nhị phân (Binary Search)

### Thiết kế thuật toán (Designed Algorithm)
---

- Input
    + $a, k, b, m, n \in N^+ $.

- Output
    + Giá trị *result* thể hiện số ngày chặt hết cây.


#### Mã giả ( Pseudo Code)

```python
def binarySearch(a, k, b, m, n):

	low = 0
	high = int(1e18)

	while low <= high:

        # Tìm vị trí chính giữa của dãy đang xét
		mid = int((low + high) // 2)

        # Tính giá trị A[mid] bằng công thức A[t] = (a+b)*t - (int(t//k)*a+int(t//m)*b)
		value = (a+b)*t - (int(t//k)*a+int(t//m)*b) 

        # Cập nhật cận trên/cận dưới.
		if value == n:
			return mid

		if value < n:
			low = mid + 1
		else:
			high = mid - 1
    
	return low
```

### Cài đặt thuật toán (Algorithm Implementation)

#### Thuật toán vét cạn

````
def bruteforce(a, k, b, m, n):
    
    number_of_tree	= 0
    day = 0
    
    while 1:
        
        day += 1

        if number_of_tree >= n:
            break
        if day % k != 0:
            number_of_tree += a
        if day % m != 0:
            number_of_tree += b

        print("Số cây chặt được cho đến ngày thứ {}: {}".format(day, number_of_tree))

    return day-1
````

````
# Test

a, k, b, m, n = map(int, input().split())

res = bruteforce(a, k, b, m, n)

print(res)
````

````
# Test với test case lớn

a, k, b, m, n = map(int, input().split())

res = bruteforce(a, k, b, m, n)

print(res)
````

#### Thuật toán tìm kiếm nhị phân

````
def binarySearch(a, k, b, m, n):

	def f(t: int):

		return (a+b)*t - (int(t//k)*a+int(t//m)*b)

	low = 0
	high = int(1e18)

	while low <= high:

		print("Low: {} --> High: {}".format(low, high))
		mid = int((low + high) // 2)

		value = f(t=mid)
		print("The middle value: {}".format(value))

		if value == n:
			return mid

		if value < n:
			low = mid + 1
		else:
			high = mid - 1

	return low
````

````
# Test

a, k, b, m, n = map(int, input().split())

res = binarySearch(a, k, b, m, n)

print(res)
````

````
# Test case lớn

a, k, b, m, n = map(int, input().split())

res = binarySearch(a, k, b, m, n)

print(res)
````
