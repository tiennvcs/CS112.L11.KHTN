## Tóm tắt bài toán

Từ 4 số nguyên $a, b, c, d$, xác định số lượng phép biến đổi (1 lần sử dụng phép biến đổi nghĩa là đồng thời tăng a và b lên 1 đơn vị) ít nhất sao cho giá trị $\frac{a}{b}$ bằng giá trị $\frac{c}{d}$, nếu không thì trả về giá trị 0.

Ví dụ 1:

$$
a = 1 \\
b = 6 \\ 
c = 2 \\
d = 3 \\
$$


- Ta có a = 1, b = 6 nên giá trị phân số $\frac{a}{b} = \frac{1}{6} \neq \frac{2}{3} = \frac{c}{d}$.
- <i>Biến đổi lần 1</i>: Tăng giá trị a và b thêm 1 đơn vị, giá trị phân số $\frac{a}{b} = \frac{1+1}{6+1} = \frac{2}{7} \neq \frac{2}{3} = \frac{c}{d}$.
- <i>Biến đổi lần 2</i>: Tăng giá trị a và b thêm 1 đơn vị, giá trị phân số $\frac{a}{b} = \frac{2+1}{7+1} = \frac{3}{8} \neq \frac{2}{3} = \frac{c}{d}$.
- <i>Biến đổi lần 3</i>: Tăng giá trị a và b thêm 1 đơn vị, giá trị phân số $\frac{a}{b} = \frac{3+1}{8+1} = \frac{4}{9} \neq \frac{2}{3} = \frac{c}{d}$.
- <i>Biến đổi lần 4</i>: Tăng giá trị a và b thêm 1 đơn vị, giá trị phân số $\frac{a}{b} = \frac{4+1}{9+1} = \frac{5}{10} = \frac{1}{2} \neq \frac{2}{3} = \frac{c}{d}$.
- <i>Biến đổi lần 5</i>: Tăng giá trị a và b thêm 1 đơn vị, giá trị phân số $\frac{a}{b} = \frac{1+1}{2+1} = \frac{2}{3} = \frac{c}{d}$


## Nhận diện khuôn mẫu

- Tham lam (Greedy)

## Thiết kế thuật toán

### Input và output

#### Input

- Dòng đầu tiên chứa giá trị số nguyên $a$.
- Dòng thứ hai chứa giá trị số nguyên $b$.
- Dòng thứ ba chứa giá trị số nguyên $c$.
- Dòng thứ tư chứa giá trị số nguyên $d$.

#### Output

- Giá trị nguyên thể hiện số lượng phép biến đổi ít nhất thỏa mãn đề bài. Nếu không thể biến đổi được thì trả về 0.

### Mã giả

```pseudo

    fucntion manipulate_equal(a, b, c, d):

        init n <- 0

        if (a - b < c - d) and (a == b or c == d):
            return 0

        while not((a == c and b == d)) or n >= 20000:
        do 
            - increase a by 1 unit.
            - increase b by 1 unit.
            - find the greatest common divisor(gcd) of a and b
            - divide a and b to gcd. 
            - increase n by 1 unit.

        if n < 200000:
            return n
        return 0
```

## Cài đặt thuật toán

````
import math
````

````
def manipulate_equal(a:int, b:int, c:int, d:int):

    n = 0

    if (a - b < c - d) and (a == b or c == d):
        return 0

    while not(a == c and b == d) or n >= 20000:
        a += 1
        b += 1
        ucln = math.gcd(a, b)
        a //= ucln
        b //= ucln
        n += 1
    if n < 200000:
        return n
    return 0
````

### Test trường hợp có thể biến đổi

````
a = int(input())
b = int(input())
c = int(input())
d = int(input())
````

````
res = manipulate_equal(a, b, c, d)
if res == 0:
    print("Can't manipulate !")
else:
    print("The minimum number of manipulations is {}".format(res))
````
