##Tóm tắt bài toán
---

Cho **N** màu lô của con đê, xác định xem có cách nào chia **N** lô này thành các bức tranh liên tiếp bắt đầu từ lô đầu tiên sao cho cuối cùng chỉ có hai loại bức tranh.\
Biết hai bức tranh giống nhau là có thể hoán vị lô của bức tranh này để giống được bức tranh kia.\
Đếm số cách để chia theo yêu cầu trên.


##Nhận diện khuôn mẫu
---

Cho dãy **a** đếm số cách chia dãy làm **K** dãy con với độ dài bằng nhau sao cho chỉ còn 2 loại dãy con khác tính chất nhau, nghĩa là tồn tại hoán vị X sao cho bằng Y.

##Cài đặt thuật toán
---

```
n = int(input().strip())
a = list(map(int, input().strip().split()))

H = [0] * (n+1)
a = [0] + a

for i in range(1,n+1):
    H[i] = H[i-1] + a[i]*(a[i]+11)*(a[i]+17)
    
def get_H(l,r):
    return H[r]-H[l-1]
    
ans = []

def solve(x, ans):
    n1 = 0
    n2 = 0
    s1 = -1 
    s2 = -1
    for i in range(1, n+1, x):
        s = get_H(i,i+x-1)
        if s==s1 or s1==-1:
            n1+=1
            s1=s
        elif s==s2 or s2==-1:
            n2+=1
            s2=s
        else: return
    if n1>0 and n2>0:
        ans += [(x,n1,n2)]

for i in range(1,n):
    if n%i==0:
        solve(i, ans)
        
if len(ans)==0:
    print(-1)
else:
    print(len(ans))
    for x in ans:
        print(*x)
```
